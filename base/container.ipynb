# 数据容器

## list
基本语法：
```python
[元素1, 元素2, 元素3, 元素4, ...]

# 定义变量
变量名 = [元素1, 元素2, 元素3, 元素4, ...]

# 定义空列表
变量名 = []
变量名 = list()
```
元素数据类型没有限制，甚至元素也可以是列表，这样就定义了嵌套列表

````python
my_list = [['Thomas', 26, True], [1, 3, 5, 7, 9]]
print(my_list)
print(type(my_list))

print(my_list[0][1])  # 通过下标访问列表中元素
print(my_list[-2][-3])  # 通过反向下标也可访问
````

````python
my_list = ['Tom', 'Jerry', 'Peter', 'July']

print(my_list.index('Jerry'))  # 指定元素值获取元素索引位置
# print(my_list.index('Thomas'))  # 如果不存在该元素则报错

my_list[0] = 'Thomas'  # 通过下标修改元素值
print(my_list)

my_list.insert(1, 'Tony')  # 通过insert在指定位置添加元素
print(my_list)

my_list.append('Jane')  # 通过append在末尾位置添加元素
print(my_list)

my_list.extend(['Kate', 'Michael'])  # 通过extend将另外容器中的元素追加到列表尾部
print(my_list)

del my_list[0]  # 通过del关键字删除列表指定下标位置的元素
print(my_list)

my_list.pop(2)  # 通过pop方法删除列表指定下标位置的元素
print(my_list)

my_list.remove('Jane')  # 通过remove方法删除元素在列表中第一个匹配项
print(my_list)

print(my_list.count('Kate'))  # 通过count方法统计元素在列表中出现次数

print(len(my_list))  # 统计列表长度

my_list.clear()  # 清空列表
print(my_list)
````

## tuple
元组一旦定义完成，就不可以修改

定义方式：
```python
(元素, 元素, 元素, ..., 元素)

# 定义元组名称
变量名称 = (元素, 元素, 元素, ..., 元素)
# 定义空元素
变量名称 = ()
变量名称 = tuple()
```

````python
t1 = (1, 2, 3)  # 使用方式与list类似，不再演示
# t1[0] = 5 # 禁止修改
````

## str

````python
my_str = "   Hello, my name is Thomas.   "

print(my_str.index("Thomas"))

my_str = my_str.replace('Thomas', 'Jerry')  # replace方法并不会修改原始字符串，而是创建出新字符串
print(my_str)

words = my_str.split(' ')
print(words)

my_str = my_str.strip()  # 去掉前后空格，同样不会修改原始字符串
print(my_str)
````

### 切片操作
序列支持切片，即：列表、元组、字符串，切片：及从一个序列中，取出一个子序列

语法：
```python
序列[起始下标:结束下标:步长]
```

````python
my_str = "   Hello, my name is T h o m a s.   "
print(my_str[21:21 + 12:2])
````

## dict
`key`不允许重复，不可以使用索引下标

基本语法：
```python
{key: value, key: value, ..., key: value}

# 定义字典变量
变量名 = {key: value, key: value, ..., key: value}
# 定义空字典
变量名 = {}
变量名 = dict()
```

````python
stu_score = {
    "Thomas": {"语文": 90, "数学": 92, "英语": 88},
    "Jerry": {"语文": 90, "数学": 92, "英语": 88},
    "Mike": {"语文": 90, "数学": 92, "英语": 88}
}
print(stu_score)
print(stu_score["Thomas"])
print(stu_score["Thomas"]['数学'])
````

````python
# 新增元素和修改元素语法相同
stu_score['Jane'] = {'语文': 60, '数学': 70, '英语': 80}
print(stu_score)

stu_score['Jane'] = {'语文': 80, '数学': 70, '英语': 80}
print(stu_score)

# 删除元素
stu_score.pop('Jane')
print(stu_score)
````

````python
# 获取所有key
keys = stu_score.keys()
print(keys)

for key in keys:
    print(stu_score[key])

print(len(stu_score))
````

````python
# 清空字典
stu_score.clear()
print(stu_score)
````
