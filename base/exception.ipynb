# 异常

捕获异常，基本语法：
```python
try:
    可能发生错误的代码
except:
    出现异常执行的代码
```

````python
try:
    print(name)
except NameError as e:
    print('name变量未定义错误')
````

````python
try:
    print(1/0)
except (NameError, ZeroDivisionError) as e:
    print(e)
````

````python
# 捕获所有异常
try:
    print(name)
except Exception as e:
    print(e)
````

````python
# else表示没有异常发生时执行的代码
try:
    print('hello world')
except Exception as e:
    print(e)
else:
    print('执行过程中未发现异常')
````

````python
try:
    f = open('test.txt', 'r')
except Exception as e:
    f = open('test.txt', 'w')
finally:
    print('执行close')
    f.close()
````
