# dict字典

## 1. 字典定义

```python
dict1 = {'张三': 71, '李四': 58, '王五': 91, '李四': 60}  # 定义有内容的字典
dict1
```

```python
dict2 = {}  # 定义无内容的字典
dict2
```

字典中的`key`必须是不可变类型，`value`可以是任意类型

```python
dict1 = {'张三': 71, '李四': 58, '王五': 91, ['唱歌', '跳舞']: 60}
```

```python
dict1 = {'张三': 71, '李四': 58, '王五': 91, ('唱歌', '跳舞'): 60}
dict1
```

## 2. 增删改查

### 2.1. 查

```python
dict1['张三']
```

```python
dict1['小夫']
```

```python
dict1.get('小夫', 'None')  # 安全取值
```

### 2.2. 增

```python
dict1['小夫'] = 80
dict1
```

### 2.3. 改

```python
dict1['小夫'] = 90  # 与新增相同，若键不存在，新增，若存在，修改
dict1
```

```python
dict1.update({'张三': 1, '李四': 8, 'Thomas': 100})  # 批量修改
dict1
```

### 2.4. 删

```python
del dict1['张三']
dict1
```

```python
dict1.pop('李四'), dict1.pop('胖虎', 'None'), dict1  # pop()删除键值对，返回值
```

```python
dict1.clear()  # 清空字典
```

## 3. 常用方法

```python
dict1 = {'大熊': 59, '小夫': 85, '静香': 91, '胖虎': 38}
dict1
```

```python
# keys()：获取所有key，dict_keys类型和list类似，但不能通过下标访问，可以通过list函数转化为list
dict1.keys(), type(dict1.keys()), list(dict1.keys())
```

```python
dict1.values()  # 获取所有的values，和keys()类似
```

```python
dict1.items()
```

## 4. 循环遍历

字典只能通过`for`循环进行遍历

```python
for k in dict1:
    print(k, dict1[k])
```
