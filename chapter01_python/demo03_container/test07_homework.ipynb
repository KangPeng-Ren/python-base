```python
print('请输入学生成绩，输入“结束”停止录入：')
score_list = []
while (True):
    score = input()
    if score == '结束':
        break
    elif not score.isnumeric():
        print('请输入正确分数格数！')
    elif not 0 <= int(score) <= 100:
        print('请输入0~100范围内的分数！')
    score_list.append(int(score))

pass_num = 0
excellent_num = 0
for item in score_list:
    if item >= 60:
        pass_num += 1
    if item >= 90:
        excellent_num += 1

print(f'总人数：{len(score_list)}')
print(f'最高分：{max(score_list)}')
print(f'最低分：{min(score_list)}')
print(f'合格人数：{pass_num}')
print(f'合格率：{((pass_num / len(score_list)) * 100):.2f}%')
print(f'优秀人数：{excellent_num}')
print(f'优秀率：{((excellent_num / len(score_list)) * 100):.2f}%')
print(f'平均分：{(sum(score_list) / len(score_list)):.2f}')
```

学生成绩列表

* 计算每名学生的平均分
* 找到总分最高的学生

```python
students = [
    {
        'name': '大熊',
        'scores': {'语文': 88, '数学': 92, '英语':95}
    },
    {
        'name': '小夫',
        'scores': {'语文': 75, '数学': 83, '英语': 80}
    },
    {
        'name': '静香',
        'scores': {'语文': 92, '数学': 95, '英语': 88}
    }
]
for stu in students:
    print(f'{stu.get('name')}均分为：{sum(stu.get('scores').values()) / len(stu.get('scores')):.2f}')

def find_best(students):
    best_stu = []
    best_score = 0
    for stu in students:
        score = sum(stu.get('scores').values())
        if score > best_score:
            best_stu.clear()
            best_stu.append(stu.get('name'))
            best_score = score
        elif score == best_score:
            best_stu.append(stu.get('name'))

    return best_stu, best_score
best_stu, best_score = find_best(students)
print(f'得分最高学生为：{best_stu}，分数为{best_score}')
```
