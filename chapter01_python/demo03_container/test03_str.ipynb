## 常用方法

```python
msg.index('o')  # 查找字符第一次出现的下标
```

```python
msg.split(' ')  # 根据分隔符分割
```

```python
msg = msg.replace('h', 'H')  # 替换字符
msg
```

```python
msg.count('o')  # 统计字符出现次数
```

```python
# strip()：从字符串中删除指定字符串中的任意字符，从字符串两端开始删除，直到遇见第一个不在指定字符串中的字符就停下
msg = ' hello world    '
msg = msg.strip()
msg
```

```python
len(msg)  # 统计字符串长度
```
