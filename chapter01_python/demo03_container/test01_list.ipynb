## 3. 增删改查

```python
list6 = []
```

### 3.1. 增

```python
for i in range(10):
    list6.append(i)  # 调用列表的.append()方法在列表末尾处新增元素
list6
```

```python
list6.insert(3, 3)  # 调用列表的.insert()方法在指定下标位置新增元素
list6.insert(7, 1)  # 调用列表的.insert()方法在指定下标位置新增元素
list6
```

```python
# 将可迭代对象中的内容依次取出，追加到列表尾部
str = 'Thomas'
list6.extend(str)
list6
```

### 3.2. 改

```python
list6[0] = 100.
```

### 3.3. 查

```python
list6[0]
```

### 3.4. 删

```python
list6.pop(3), list6  # 调用列表的.pop()方法弹出指定下标位置的元素，并将其返回
```

```python
list6.remove(1)  # 调用列表的.remove()方法删除列表中第一次出现的值
list6
```

```python
del list6[3]  # 通过del关键字删除列表指定位置的元素
list6
```

```python
list6.clear()  # 调用列表的.clear()方法清空列表
list6
```

## 4. 常用方法

```python
list7 = ['banana', 'apple', 'orange', 'apple']
```

```python
list7.index('banana')  # 查找指定元素在列表中第一次出现的位置
```

```python
list7.count('banana')  # 统计元素在列表中出现次数
```

```python
list7.reverse()  # 反转列表
list7
```

## 5. 容器的内置函数（适用所有容器）

```python
list8 = [4, 2, 7, 3, 9, 1, 7, 6, 3, 5, 8]
list8
```

```python
result = sorted(list8, reverse=False)  # 调用sorted()函数对容器进行排序，返回新容器
result, id(result), id(list8), type(result)
```

```python
len(list8)  # 调用len()函数统计容器中元素个数并返回
```

```python
max(list8), max(33, 45, 88, 9)  # 调用max()函数返回容器中最大元素
```

```python
min(list8), min(33, 45, 88, 9)  # 调用min()函数返回容器中最小元素
```

```python
sum(list8)  # 对容器中所有元素求和（不适用字符串），并且要求容器中所有元素为数值类型
```

## 6. 循环遍历

```python
list9 = [62, 50, 60, 48, 80, 20, 95]
```

```python
idx = 0
while idx < len(list9):
    print(list9[idx])
    idx += 1
```

```python
for item in list9:
    print(item)

for idx in range(len(list9)):
    print(list9[idx])

for idx, item in enumerate(list9):
    print(idx, item)
```
