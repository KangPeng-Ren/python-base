# 多元交叉熵损失函数

```python
import torch
import torch.nn as nn
```

```python
# 1. 目标值为顺序编码的标签
input = torch.randn(6, 8)
print(input)

pred = torch.softmax(input, dim=1)
print(pred)
```

```python
# 目标值，6个数据的标签类别（0~7）
target = torch.tensor([1, 0, 3, 7, 5, 2])
print(target)
```

```python
# 定义交叉熵损失函数
loss = nn.CrossEntropyLoss()
print(loss(pred, target))
print(loss(input, target))  # 不适用softmax激活，直接通过原始数据也可以计算损失函数
```

```python
# 2. 目标值设置为独热编码（一组概率）
input = torch.randn(6, 8)
print(input)
```

```python
target = torch.randn(6, 8).softmax(dim=1)
print(target)
```

```python
loss = nn.CrossEntropyLoss()
print(loss(input, target))
```
