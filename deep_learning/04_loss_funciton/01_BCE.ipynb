# 交叉熵损失函数

```python
import torch
import torch.nn as nn
```

```python
# 输入数据
input = torch.randn(3, 2)
print(input)
# 得到预测概率值
pred = torch.sigmoid(input)
print(pred)
```

```python
# 目标值
target = torch.tensor([[0, 1], [1, 0], [0, 1]], dtype=torch.float)
print(target)
```

```python
# 定义损失函数
loss = nn.BCELoss()
print(loss(pred, target))
```
