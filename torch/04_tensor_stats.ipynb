```python
import torch
```

# 张量统计相关

```python
tensor1 = torch.randint(1, 10, (3, 2, 4)).float()
print(tensor1)
```

```python
# sum 求和
print(tensor1.sum())
print(tensor1.sum(dim=0))  # 降第0维
print(tensor1.sum(dim=1))
print(tensor1.sum(dim=2))
```

```python
# mean 求均值，要求数据类型必须为float型或者Complex型
print(tensor1.mean())
print(tensor1.mean(dim=2))
```

```python
# std 求标准方差
print(tensor1.std())
print(tensor1.std(dim=2))
```

```python
# max min 求最大最小值
print(tensor1.max())
print(tensor1.max(dim=2))  # 指定维度，可以同时获取该维度最值和其对应的索引
print(tensor1.min())
print(tensor1.min(dim=2))

print(tensor1.argmin())
```

```python
# unique 去重
print(tensor1.unique())  # 得到一维张量
print(tensor1.unique(dim = 2))
```

```python
# sort 排序
print(tensor1.sort())   # 在这里默认 dim=2
```
