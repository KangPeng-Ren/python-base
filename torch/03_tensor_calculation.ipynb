```python
import torch
```

# 张量的运算

```python
tensor1 = torch.randint(low=0, high=10, size=(2, 3))
print(tensor1)
```

```python
print(tensor1 + 10)
print(tensor1)
```

```python
print(tensor1.add(10))  # add()方法并不会修改原张量中的值
print(tensor1)
```

```python
print(tensor1.add_(10))  # add_()方法会修改原张量中的值，同理，sub() multi() div() neg() 相同
print(tensor1)
```

```python
print(tensor1.sub(10).pow(3))  # pow_()方法会修改原张量中的值

print(tensor1.pow_(2))
print(tensor1.sqrt())
```

```python
# 哈达玛积
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[1, 2], [3, 4]])
print(tensor1 * tensor2)
print(tensor1.mul(tensor2))
```

```python
# 二维矩阵乘法
tensor1 = torch.tensor([[1, 2, 3], [4, 5, 6]])
tensor2 = torch.tensor([[1, 2], [3, 4], [5, 6]])
# mm()严格用于二维矩阵乘法
print(tensor1.mm(tensor2))
# @ matmul()支持多维矩阵，按最后两个维度做矩阵乘法，其他维度相同，或者至少为1个张量维度为1，广播后进行运算
print(tensor1 @ tensor2)
print(tensor1.matmul(tensor2))
```
