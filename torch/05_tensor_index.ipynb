```python
import torch
```

# 张量索引相关

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
# 简单索引
print(tensor1[2, 1, 3])
print(tensor1[2, 3])
```

```python
# 范围索引
print(tensor1[:, 3])
print(tensor1[1:])
print(tensor1[:, 2:4, 3])
```

```python
# 列表索引
print(tensor1[[0, 1], [1, 2], [1, 3]])  # 所选取的元素坐标
print(tensor1[[[0], [1]], [1, 2]])  # 一对多的关系
```

```python
# 布尔索引
mask = tensor1[:, :, 0] > 5  # 判断每行首元素是否大于5
print(mask)
print(tensor1[mask])
```

```python
mask = tensor1[:, 1, :] > 5  # 判断每列第二个元素是否大于5
print(mask)
tensor2 = tensor1.permute(0, 2, 1)  # permute：一般化的转置
print(tensor2[mask])
```

```python
# 选取符合条件的矩阵
mask = tensor1[:, 0, 1] > 7
print(mask)
print(tensor1[mask])
```
