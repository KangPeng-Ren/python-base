```python
import torch
```

# 张量形状操作

```python
tensor1 = torch.randint(1, 20, (2, 3, 6))
print(tensor1)
```

```python
# 交换两个维度
print(tensor1.transpose(1, 2))
```

```python
# 重新排列多个维度
print(tensor1.permute(2, 1, 0))
```

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
# 调整形状
print(tensor1.reshape((6, 10)))  # 各维度相乘要与原张量各维度相乘结果相同
```

```python
# view() 调整张量的形状，需要内存连续，共享内存
print(tensor1.is_contiguous())  # 判断张量是否内存连续，若是将张量进行反转，则会造成不连续
# tensor1.contiguous()  # 将张量转化为内存连续
print(tensor1.view(-1, 10))
print(tensor1.shape)
```

```python
# 增加或删除维度，最后一个维度为-1维度，同样 unsqueeze_()方法可以原地操作
tensor1 = torch.tensor([1, 2, 3, 4, 5])
# print(tensor1.unsqueeze(0)) # 在0维度上添加一个维度
print(tensor1.unsqueeze_(1)) # 在维度1上添加一个维度

print(tensor1.squeeze(1))   # 要求对维度为1的维度进行压缩
```
