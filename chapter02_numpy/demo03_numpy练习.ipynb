某城市一周的最高气温（℃）为`[28, 30, 29, 31, 32, 30, 29]`。

- 计算平均气温、最高气温和最低气温。
- 找出气温超过30℃的天数。

```python
temperature_arr = np.array([28, 30, 29, 31, 32, 30, 29])
print(
    f"平均气温：{np.mean(temperature_arr):.2f}℃，最高气温：{np.max(temperature_arr)}℃，最低气温：{np.min(temperature_arr)}℃")
print(f"超过30℃的天数有：{len(temperature_arr[temperature_arr > 30])}")
```

学生成绩统计：班级5名学生的数学成绩为`[85, 90, 78, 92, 88]`。

- 计算成绩的平均分、中位数和标准差。
- 将成绩转换为十分制（假设满分为100）。

```python
scores_list = np.array([85, 90, 78, 92, 88])
print(f"平均分：{np.mean(scores_list)}，中位数：{np.median(scores_list)}，标准差：{np.std(scores_list)}")
print(f"十分制：{scores_list / 10}")
```

给定矩阵 `A=[[1, 2], [3, 4]]` 和 `B=[[5, 6], [7, 8]]`。

* 计算A+B和A*B（逐元素乘法）。
* 计算A和B 的矩阵乘法（点积）。

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(f"A + B = {A + B}")
print(f"A * B = {A * B}")
print(f"A @ B = {A @ B}")
```

随机数据生成：生成一个`(3, 4)`的随机整数数组，范围`[0，10)`。

- 计算每列的最大值和每行的最小值。
- 将数组中的所有奇数替换为-1

```python
arr = np.random.randint(low=0, high=10, size=(3, 4))
print(arr)
print(f"每列的最大值为：{np.max(arr, axis=0)}")
print(f"每行的最小值为：{np.min(arr, axis=1)}")
# print(f"将所有奇数换为-1后：{np.where(arr % 2 == 0, arr, -1)}")
arr[arr % 2 == 1] = -1
print(f"将所有奇数换为-1后：{arr}")
```

数组变形：创建一个 1 到 12 的一维数组，并转换为(3, 4)的二维数组

- 计算每行的和与每列的平均值。
- 将数组展平为一维数组。

```python
arr1 = np.arange(start=1, stop=13, step=1)
arr2 = arr1.reshape(3, 4)
print(f"arr1 = {arr1}")
print(f"arr2 = {arr2}")
print(f"行和：{np.sum(arr2, axis=1)}")
print(f"列平均：{np.mean(arr2, axis=0)}")
print(f"展平为一维数组：{arr2.reshape(12)}")
```

布尔索引：生成一个`(5, 5)`的随机数组，范围[0，20)。

- 找出数组中大于10的元素。
- 将所有大于10的元素替换为0。

```python
arr = np.random.randint(low=0, high=20, size=(5, 5))
print(arr)
print(f"数组中大于10的元素：{arr[arr > 10]}")
arr[arr > 10] = 0
print(f"将数组中大于10的元素替换为0：{arr}")
```

统计函数应用：某公司6个月的销售额（万元）为`[120, 135, 110, 125, 130, 140]`。

- 计算销售额的总和、均值和方差。
- 找出销售额最高的月份和最低的月份

```python
sales = np.array([120, 135, 110, 125, 130, 140])
print(f"销售总和：{np.sum(sales)}，均值：{np.mean(sales):.2f}，方差：{np.var(sales):.2f}")
print(f"销售额最高的月份为：{np.argmax(sales) + 1}，最低的月份为：：{np.argmin(sales) + 1}")
```

数组拼接：给定`A=[1, 2, 3]`和`B=[4, 5, 6]`。

- 将 A 和 B 水平拼接为一个新数组。
- 将 A 和 B 垂直拼接为一个新数组。

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
print(f"A和B水平拼接后：{np.concatenate([A, B])}")
print(f"A和B垂直拼接后：{np.concatenate([A, B]).reshape(2, 3)}")
```

唯一值与排序：给定数组`[2, 1, 2, 3, 1, 4, 3]`。

- 找出数组中的唯一值并排序。
- 计算每个唯一值出现的次数

```python
arr = np.array([2, 1, 2, 3, 1, 4, 3])
print(f"去重排序：{np.unique(arr)}")
u_arr, counts = np.unique(arr, return_counts=True)
print(f"{u_arr}出现次数为：{counts}")
```

某商店5天的销售额（万元）和成本（万元）如下：

销售额：`[20, 25, 22, 30, 28]`
成本：`[15, 18, 16, 22, 20]`

- 计算每天的利润（销售额－成本）。
- 计算利润的平均值和标准差。
- 找出利润最高的天数。

```python
sales = np.array([20, 25, 22, 30, 28])
cost = np.array([15, 18, 16, 22, 20])
profit = sales - cost
print(f"利润：{profit}")
print(f"利润平均值：{np.mean(profit)}，标准差：{np.std(profit):.2f}")
print(f"利润最高的天数为：{len(profit[profit==np.max(profit)])}")
```
