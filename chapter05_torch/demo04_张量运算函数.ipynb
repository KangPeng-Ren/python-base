# 张量运算函数

```python
import torch

tensor1 = torch.randint(low=1, high=9, size=(3, 2, 4))
tensor1
```

* `sum()`求和

```python
tensor1.sum(), tensor1.sum(dim=(0, 1))  # 通过dim参数指定求和的维度
```

* `mean()`求均值

```python
tensor1 = tensor1.float()
tensor1.mean(), tensor1.mean(dim=(0, 1))
```

* `max()/min()`求最大值/最小值及其索引

```python
tensor1.max(), tensor1.max(dim=1), tensor1.min(), tensor1.min(dim=1)
```

* `argmax()/argmin()`求最大值/最小值的索引

```python
tensor1.argmax(), tensor1.argmax(dim=1), tensor1.argmin(), tensor1.argmin(dim=1)
```

* `std()`求标准差

```python
tensor1.std()
```

* `unique()`去重

```python
tensor1.unique()
```

`sort()`排序

```python
tensor1.sort()
```
