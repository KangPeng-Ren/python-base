# 张量形状操作

```python
import torch

tensor1 = torch.randint(1, 9, (2, 3, 6))
tensor1
```

## 1. 交换维度

* `transpose()`交换两个维度

```python
tensor1.transpose(1, 2)  # 交换第1维和第2维
```

* `permute()`重排列多个维度

```python
tensor1.permute(2, 0, 1)  # (2, 3, 6) -> (6, 2, 3)
```

# 2. 调整形状

* `reshape()`调整张量维度

```python
tensor1.reshape(6, 6), tensor1.reshape((12, 3))
```

* `view()`调账张量形状，需要内存连续，内存共享
* `is_contiguous()`判断是否内存连续
* `coutiguous()`转化维连续内存

```python
tensor1 = tensor1.reshape(6, 6)
print(tensor1, tensor1.is_contiguous())
print(tensor1.view(-1, 12))

tensor2 = tensor1.T
print(tensor2, tensor2.is_contiguous())
tensor2.contiguous()
print(tensor2.contiguous().view(-1, 12))
```

## 3. 增加或删除维度

* `unsqueeze()`在指定维度上增加1个维度

```python
tensor1 = torch.tensor([1, 2, 3, 4, 5])
print(tensor1)

print(tensor1.unsqueeze(dim=0)) # 在第0维上增加一个维度
print(tensor1.unsqueeze(dim=1)) # 在第1维上增加一个维度
```

* `squeeze()`删除大小为1的维度

```python
tensor1 = torch.tensor([1, 2, 3, 4, 5])
print(tensor1.unsqueeze_(dim=1))
print(tensor1.squeeze())
```
