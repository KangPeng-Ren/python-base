# 张量拼接操作

```python
import torch

tensor1 = torch.randint(1, 9, (2, 2, 5))
tensor2 = torch.randint(1, 9, (2, 2, 5))
tensor1, tensor2
```

* `torch.cat()`张量拼接，按已有维度拼接。除拼接维度外，其他维度大小必须相同

```python
torch.cat([tensor1, tensor2], dim=1)
```

* `torch.stack()`张量堆叠，按新维度堆叠，所有张量形状必须一致

```python
tensor3 = torch.stack([tensor1, tensor2], dim=2)
tensor3, tensor3.shape
```
